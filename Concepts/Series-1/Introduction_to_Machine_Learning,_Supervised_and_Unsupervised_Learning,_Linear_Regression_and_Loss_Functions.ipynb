#**Supervised and Unsupervised learning**

Before we dive into linear regression, let's talk about learning. At the end of the day, Machine learning takes input data and makes a prediction based on the data passed in.

There are mainly two types of ways we can pass in data to train our model, and they are **Supervised** and **Unsupervised** learning.

## **Supervised Learning**:


Supervised learning is used to train our ML model with a labeled dataset, this is also called a **training set**. A **labeled dataset** means that the input variables **(features)** also contain the correct corresponding labeled outputs.

For example, if we want to predict house prices based on the size (sq ft) of the home, the data we pass into our model is labeled as price and size, as shown below:

| Size (Sq Ft) | Price (in thousands of dollars)
| :-: | :-: |
| 1100 | 700
| 2135 | 1400
| 800 | 500
| 1200 | 600
| 1500 | 900
| 1700 | 1000
| ... | ...

In this example, the input would be the size of the home and our model would try to predict the price of the home based off the size.

In Supervised learning, the idea is that we feed the training set into our ML algorithm to train the model and output a prediction. It does this continuously to become better at predicting house prices. In the world of machine learning, we can also call the output of our algorithm as the **hypothesis (h)**.

Linear regression and classification are two types of algorithms used with supervised learning, but we will get into those later.



## **Unsupervised Learning**:

Unlike supervised learning, unsupervised learning doesn't have labels on the dataset that is used to train the model. This means that there are **no outputs** in our training data. The dataset only contains input data, and our model will analyze the data and find underlying patterns, groupings, or any type of associations with that data.

A good example is to imagine you have a large set of customer data, including purchase history, browsing patterns, and demographics. In unsupervised learning, you might use clustering algorithms to segment customers into distinct groups (nodes) based on their behavior, without prior knowledge of what these groups should be. This is known as **Clustering**.

Another example is we can use unsupervised learning to determine what input features are important and which input features can be ignored, to reduce the complexity of our dataset. This is known as **Dimensionality Reduction**.

Unsupervised learning is great for segmenting data into different groups (clustering), discovering hidden patterns, compressing data (dimensionality reduction), and detecting anomalies in our data.

#**Linear Regression**

Linear regression is one of the most simple and fundamental algorithms in machine learning. We use linear regression to find a correlation (or line of best fit) between an independent and dependent variable. As we mentioned before, since our dataset for training a regression model contains the input and output data, linear regression is a form of **supervised learning**.

The linear regression equation is almost the same as the line equation but there are a few slight variations:

###**h(x) = $\theta_0$ + $\theta_1$x**

where **h** is our hypothesis (output), **x** is our input, and **$\theta$** represents our parameters (more on this later).

If we go back to our example with home prices in the previous section, let's say we added another feature (variable) to our dataset and that is the number of bedrooms. Our dataset would now look like this:

| Size (Sq Ft) | No. of Bedroooms | Price (in thousands of dollars)
| :-: | :--: | :-: |
| 1100 | 3 | 700
| 2135 |  5 | 1400
| 800 | 2 | 500
| 1200 | 3 | 600
| 1500 | 3 | 900
| 1700 | 2 | 1000
| ... | ... | ...

Now we have multiple input features in our training set, so we can adjust our regression equation to become:

###**$h_\theta$($x_1$,$x_2$) = $\theta_0$ + $\theta_1$$x_1$ + $\theta_2$$x_2$**  

In this case, $\theta$ becomes a three dimensional parameter and can be written in a matrix as

\begin{bmatrix} \theta_0 \\\ \theta_1 \\\ \theta_2 \end{bmatrix}

and the features **x** can be written as

\begin{bmatrix} x_0 \\\ x_1 \\\ x_2 \end{bmatrix}

where $x_0$ = 1. Generally, if we have **n** input features, then our matricies will have **n + 1** dimensions, since we include the 0.

I noted that $\theta$ represents our parameters, and from the equation we can see that for each input feature, we multiply it with a parameter. **The job of the learning algorithm is to choose the values for the parameters that improve the prediction accuracy.**

##**Testing how accurate our model is**

In the world of machine learning, we can use **loss functions** to calculate how accurate or "good" our model is. Loss functions take the predicted outputs from our model (i.e hypothesis) and compare them with the actual outputs to see how accurate the model is. The goal is to minimize our loss functions. When we design our machine learning algorithms to minimize the loss functions, that means there is less error in predicting an output. In Linear Regression, we can use the **Mean Squared Error (MSE)** function:


###**MSE = $\frac{1}{2n}$$\sum_{i=1}^n (h_\theta(x_i)-y_i)^2$**

where **n** is the number of training examples (i.e the number of rows in the table from our example), **i**  represents the **ith** training example, $h_\theta(x_i)$ represents the predicted output from our model when we pass the **ith** training example in, and $y_i$ represents the actual output from the **ith** training example.

We can also define our loss function as J($\theta$):

###**J($\theta$) = $\frac{1}{2n}$$\sum_{i=1}^n (h_\theta(x_i)-y_i)^2$**

**Note** that for regression we use the MSE loss function, but this can change depending on the type of algorithm we use. In later lessons we will take a deeper look into the different loss functions.
